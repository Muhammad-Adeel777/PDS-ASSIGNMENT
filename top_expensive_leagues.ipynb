```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("top_expensive_leagues.csv")
```

```python
# View the first few rows
print(df.head())
```

```python
# View the last few rows
print(df.tail())
```

```python
# Get statistical summaries
print(df.describe())
```

```python
# Count unique values in a column
print(df['Country'].value_counts())
```

```python
# Get unique values in a column
print(df['League Name'].unique())
```

```python
# Check for missing values
print(df.isnull())
```

```python
# Fill missing values (example: fill with 0)
df.fillna(0, inplace=True)
```

```python
# Drop rows with missing data
df.dropna(inplace=True)
```

```python
# Convert a column to datetime (example for a 'Founded Year' column)
df['Founded Year'] = pd.to_datetime(df['Founded Year'], format='%Y')
```

```python

# Drop unnecessary columns (example: dropping 'Top Team')
df.drop(columns=['Top Team'], inplace=True)
```

```python

# Reset the index after transformations
df.reset_index(drop=True, inplace=True)
```

```python
# Access specific rows and columns
print(df.loc[0:2, ['League Name', 'Country']])
```

```python
# Convert a column into a NumPy array
revenue_array = np.array(df['Revenue (USD)'])
revenue_array
```

```python

# Create a sequence of numbers (example: years from 1900 to 2025)
years = np.arange(1900, 2026, 5)
years
```

```python

# Perform element-wise operations (example: double revenue)
doubled_revenue = revenue_array * 2
doubled_revenue
```

```python

# Get array properties
print(revenue_array.ndim)  # Dimension
print(revenue_array.shape)  # Shape
```

```python
# Manual entries for better visuals
league_names = ['Premier League', 'Serie A', 'IPL']  # League names
revenues = [5275330000, 7088640000, 7930520000]  # Corresponding revenue values

# Plotting the line plot
plt.plot(league_names, revenues, marker='o', linestyle='--', color='purple', linewidth=2, markersize=8)

# Customizing the axes and title
plt.xlabel('League Name', fontsize=12)
plt.ylabel('Revenue (USD)', fontsize=12)
plt.title('Revenue by League', fontsize=14, fontweight='bold')

# Adding grid for better readability
plt.grid(True, linestyle='--', alpha=0.6)

# Display the plot
plt.show()

```

```python
# Bar chart of revenue by league
plt.bar(df['League Name'], df['Revenue (USD)'], color='blue')
plt.xlabel('League Name')
plt.ylabel('Revenue (USD)')
plt.title('Revenue by League')
plt.show()
```

```python
# Scatter plot of revenue vs viewership
plt.scatter(df['Revenue (USD)'], df['Viewership'], color='green')
plt.xlabel('Revenue (USD)')
plt.ylabel('Viewership')
plt.title('Revenue vs Viewership')
plt.show()
```

```python
# Manual entries for better visuals
slices = [5275330000, 7088640000, 7930520000]  # Revenue values (adjust these manually)
labels = ['Premier League', 'Serie A', 'IPL']  # League names
colors = ['gold', 'lightblue', 'lightgreen']   # Custom colors for each slice
explode = (0.1, 0, 0.1)  # Explode the first and last slices

# Plotting the pie chart
plt.pie(
    slices,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',
    startangle=90,
    explode=explode
)
plt.title('Revenue Distribution by League')
plt.show()
```

```python
# Save a bar chart as an image
plt.bar(df['League Name'], df['Revenue (USD)'], color='blue')
plt.xlabel('League Name')
plt.ylabel('Revenue (USD)')
plt.title('Revenue by League')
plt.savefig('revenue_by_league.png')
plt.show()
```
